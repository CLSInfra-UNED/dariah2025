```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Simulación de tus datos (puedes reemplazar esto por tu CSV real)
df = pd.DataFrame({
    'prompt': ['A', 'A', 'B', 'B', 'C', 'C'] * 10,
    'score': [1, 2, 2, 3, 3, 4] * 10
})

# Limpieza (por si acaso hay espacios u otros problemas)
df['prompt'] = df['prompt'].astype(str).str.strip()
df = df.dropna(subset=['prompt', 'score'])

# Convertir 'prompt' a categoría explícita
df['prompt'] = df['prompt'].astype('category')

# Graficar
sns.histplot(data=df, x='score', hue='prompt', kde=True, element='step')

# Añadir leyenda y título
plt.title('Distribución por Prompt')
plt.xlabel('Score')
plt.ylabel('Frecuencia')
plt.legend(title='Prompt')  # Esto DEBE mostrar los valores A, B, C con colores
plt.show()
```
